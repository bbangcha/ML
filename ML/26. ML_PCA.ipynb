### numpy의 RandomState 
- 클래스는 난수 생성기(Random Number Generator, RNG)를 포함하는 클래스
- 이 클래스는 무작위 수를 생성할 때 사용되며, 초기 시드(seed) 값을 지정하여 생성되는 난수를 재현 가능
- rand(d0, d1, ..., dn) : 0에서 1 사이의 균일 분포에서 난수를 생성
- randn(d0, d1, ..., dn) : 평균이 0이고 표준 편차가 1인 표준 정규 분포에서 난수를 생성

### numpy의 dot
- 함수는 두 배열의 내적(dot product)을 계산하는 데 사용
- 내적은 두 벡터의 성분별 곱셈 후 그 결과를 모두 더하는 연산이며, 두 행렬 간의 행렬 곱셈도 가능
- 기본적으로 numpy.dot(a, b)는 두 배열 a와 b의 내적을 반환하며 a와 b는 1차원 또는 2차원 배열이어야 가능

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('white')

rng = np.random.RandomState(13)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
# .T : .transpose 와 같은 의미로 전치 즉, 배열이나 행렬에서의 행과 열을 바꾸는 것을 의미
X.shape
```

### PCA
- 차원축소와 변수추출 기법으로 쓰이고 있는 주성분분석
- 변수추출 : 데이터의 분산을 최대한 보존하면서 기존 변수를 조합해 새로운 변수를 만드는 기법
- 여러 변수 중에서 특수한 특성을 가지는 변수들만의 관계를 보는 변수 선택과는 다름
- n_components : 표현할 주성분 갯수, 여기선 2개의 주성분으로 표현하는 설정

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2, random_state = 13)
pca.fit(X)
```

### 벡터값과 분산값(=설명력)

```python
#벡터값
pca.components_
```

```python
# 각 벡터의 분산값
pca.explained_variance_
```

```python
# 각 벡터의 분산값(=설명력)의 비율
pca.explained_variance_ratio_
```

```python
# 좌표의 원점, 데이터의 중심, 가장 큰 영향력을 끼치는 벡터
pca.mean_
```

### 주성분 벡터 그리기

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops = dict(arrowstyle= '->',
                      linewidth = 2, color = 'black',
                      shrinkA = 0, shrinkB = 0)
    ax.annotate('', v1, v0, arrowprops = arrowprops)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.4)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal')
plt.show()
```

### PCA.transform을 통하여 X데이터를 위 PCA에서 설정한 한개의 주성분인(=1차원인) 데이터로 변경

```python
X_pca = pca.transform(X)
X_pca
```

### X데이터를 1차원으로 만든 X_pca를 그래프로 표현
- 원래 데이터가 2차원이기 떄문에 .inverse_transform 이용하여 X_pca를 다시 2차원을 변환 후 그래프로 표현

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:,0], X[:,1], alpha=0.4)
plt.scatter(X_new[:,0], X_new[:,1], alpha=0.8)
plt.show()
```

### iris 데이터

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

iris_pd = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_pd['species'] = iris.target
iris_pd.head()
```

```python
sns.pairplot(iris_pd, hue='species', height=3,
             x_vars=['sepal length (cm)', 'petal width (cm)'],
             y_vars=['petal length (cm)', 'sepal width (cm)'])
```

### Scaler 적용

```python
from sklearn.preprocessing import StandardScaler

iris_ss = StandardScaler().fit_transform(iris.data)
iris_ss[:3]
```

### PCA 결과 출력
- n_components = 2 : 주성분을 2개로 설정

```python
from sklearn.decomposition import PCA

def get_pca_data(ss_data, n_components = 2):
    pca = PCA(n_components=n_components)
    pca.fit(ss_data)

    return pca.transform(ss_data), pca
```

```python
iris_pca, pca = get_pca_data(iris_ss, 2)
iris_pca.shape
```

```python
pca.mean_
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

### PCA 결과 pandas 정리 및 시각화

```python
def get_pd_from_pca(pca_data, cols=['pca_component_1', 'pca_component_2']):
    return pd.DataFrame(pca_data, columns=cols)
```

```python
iris_pd_pca = get_pd_from_pca(iris_pca)
iris_pd_pca['species'] = iris.target
iris_pd_pca.head()
```

```python
sns.pairplot(iris_pd_pca, hue='species', height=5,
             x_vars=['pca_component_1'], y_vars=['pca_component_2'])
```

##### 교차검증 한번에 하기 : cross_val_score
- cross_val_score(estimator, feature, label, scoring = , cv=)
- scoring = : 예측 성능 평가 지표 (ex. accurency)
- cv = : 데이터를 나눈 수, 교차 검증 폴드 수

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

def rf_scores(X, y, cv=5):
    rf = RandomForestClassifier(n_estimators=100, random_state=13)
    scores_rf = cross_val_score(rf, X, y, scoring = 'accuracy', cv=cv)

    print('Score : ', np.mean(scores_rf))

rf_scores(iris_ss, iris.target)
```

```python
pca_X = iris_pd_pca[['pca_component_1', 'pca_component_2']]
rf_scores(pca_X, iris.target)
```

### wine 데이터

```python
import pandas as pd

wine_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/wine.csv'

wine = pd.read_csv(wine_url, index_col=0)
# index_col=0 : csv파일을 불러올 시 새로운 컬럼(=unnamed : 0)을 생성시키지 않고 파일을 불러오겠다는 명령어
wine.head()
```

```python
wine_y = wine['color']
wine_X = wine.drop(['color'], axis=1)
wine_X.head()
```

### StandardScaler 적용

```python
wine_ss = StandardScaler().fit_transform(wine_X)
wine_ss[:3]
```

```python
pca_wine, pca = get_pca_data(wine_ss, n_components=2)
pca_wine.shape
```

#### 2가지 주성분으로 나눌 시 설명력(=분사값)이 50%도 되지 않는다 

```python
def print_variance_ration(pca):
    print('variance_ration : ', pca.explained_variance_ratio_)
    print('sum of variance_ration : ', np.sum(pca.explained_variance_ratio_))
```

```python
print_variance_ration(pca)
```

```python
pca_columns = ['pca_component_1', 'pca_component_2']
pca_wine_pd = pd.DataFrame(pca_wine, columns=pca_columns)
pca_wine_pd['color'] = wine_y.values

sns.pairplot(pca_wine_pd, hue='color', height = 5,
             x_vars=['pca_component_1'], y_vars=['pca_component_2'])
```

### RandomForest 적용 시 원 데이터와 큰 차이가 없다

```python
# wine_ss : StandardScaler를 적용한 원데이터
rf_scores(wine_ss, wine_y)
```

```python
# pca로 주성분을 2개로 만든 데이터
pca_X = pca_wine_pd[['pca_component_1', 'pca_component_2']]
rf_scores(pca_X, wine_y)
```

### 주성분 3개로 조정
- 98% 이상 표현 가능

```python
pca_wine, pca = get_pca_data(wine_ss, n_components=3)

print_variance_ration(pca)
```

```python
cols = ['pca_1', 'pca_2', 'pca_3']
pca_wine_pd = get_pd_from_pca(pca_wine, cols=cols)

pca_X = pca_wine_pd[cols]
rf_scores(pca_X, wine_y)
```

### 3D로 표현

```python
pca_wine_plot = pca_X
pca_wine_plot['color'] = wine_y.values
pca_wine_plot
```

```python
from mpl_toolkits.mplot3d import Axes3D

markers = ['^', 'o']

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for i, markers in enumerate(markers):
    x_axis_data = pca_wine_plot[pca_wine_plot['color']==i]['pca_1']
    y_axis_data = pca_wine_plot[pca_wine_plot['color']==i]['pca_2']
    z_axis_data = pca_wine_plot[pca_wine_plot['color']==i]['pca_3']

    ax.scatter(x_axis_data, y_axis_data, z_axis_data,
               s = 20, alpha = 0.5, marker = markers)
    

ax.view_init(30, 80)
plt.show()
```
