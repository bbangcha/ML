### 다항함수 그리기 $f(x) = 3x^2 + 2$
- 3 = 게수
- x = 변수
- ^2 = 차수
- +2 = 상수

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

x = np.linspace(-3, 2, 100)
# linespace(시작, 끝, 만들숫자개수)
y = 3 * x ** 2 + 2

mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.xlabel('$x$', fontsize = 10)
            # $ 기호 싸아주면 수학기호로 변경
plt.ylabel('$y$', fontsize = 10)
plt.show()
```

### X축 이동 $f(x) = 3(x+1)^2 + 2$
- X축을 기준으로 -1 만큼 평행이동

```python
x = np.linspace(-3, 2, 100)
# linespace(시작, 끝, 만들숫자개수)
y1 = 3 * x ** 2 + 2
y2 = 3 * (x + 1) ** 2 + 2
```

```python
mpl.style.use('seaborn-whitegrid')

# $ 기호 싸아주면 수학기호로 변경
plt.figure(figsize=(6, 4))
plt.plot(x, y1, lw = 2, ls = 'dashed', label = '$f(x) = 3x^2 + 2$') 
plt.plot(x, y2, label = '$f(x) = 3(x+1)^2 + 2$')
plt.legend(fontsize = 10)
plt.xlabel('$x$', fontsize = 10)
plt.ylabel('$y$', fontsize = 10)
plt.show()
```

### 지수함수 $f(x) = a^x$
- a = 밑수
- x = 지수

```python
x = np.linspace(-2, 2, 100)
a11, a12, a13 = 2, 3, 4
y11, y12, y13 = a11**x, a12**x, a13**x

a21, a22, a23 = 1/2, 1/3, 1/4
y21, y22, y23 = a21**x, a22**x, a23**x
```

### 밑수의 변화에 따른 지수함수 변화
- $(1/2)^x$ = $(2^{-1})^ x$ = $2^{-x}$ 이므로 $2^x$ 과 대칭을 이룬다
- 지수가 정수이면서 점점 커지게 되면 급격하게 증가하는 형태를 띈다
- 지수가 분수이면서 점점 작아지면 (=지수가 정수이고 -x 승이면서 점점 커지면) 급격하게 0으로 감소하는 형태는 띈다

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(x, y11, color='k', label=r'$2^x$')
ax[0].plot(x, y12, '--', label=r'$3^x$')
ax[0].plot(x, y13, ':', label=r'$4^x$')
ax[0].legend(fontsize=10)

ax[1].plot(x, y21, color='k', label=r'$(1/2)^x$')
ax[1].plot(x, y22, '--', label=r'$(1/3)^x$')
ax[1].plot(x, y23, ':', label=r'$(1/4)^x$')
ax[1].legend(fontsize=10)

plt.show()
```

### 지수 증가의 위력
- 지수 증가는 급격하게 커지는 경향을 띈다

```python
x = np.linspace(0, 10)

plt.figure(figsize=(4, 4))

plt.plot(x, x**2, '--', color='k', label=r'$x^2$')
plt.plot(x, 2**x, color='k', label=r'$2^x$')

plt.legend(loc='center left', fontsize=10)
plt.xlabel('$x$', fontsize=10)
plt.ylabel('$y$', fontsize=10)

plt.show()
```

### 로그함수 : $f(x) = log_ax$
- $2^x$ = 16 -> $log_2 16$ = 4
- 로그의 특성 : $log_a x$ = $logx$/$loga$
- $e$ = 2.71828..
- $log_2$ > $log_{e}$ > $log_{10}$ : 로그의 밑수가 커지면 전체 log값은 작아진다


```python
def log(x, base):
    return np.log(x)/np.log(base)
    # 마크 다운의 로그 특성을 이용하여 함수로 만들고 적용

x1 = np.linspace(0.001, 5, 1000)
x2 = np.linspace(0.01, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)
y21, y22 = log(x1, 1/10), log(x2, 1/np.e)
```

### 로그함수의 특징

- 밑수(ex.10, e)가 정수이고 작으면 가파르게 올라가지만 수렴하는 값은 작다
- 밑수가 정수이고 크면 완만하게 올라가지만 수렴하는 값은 크다
- 밑수가 분수이면서 크면 (=지수가 정수이고 - 승이면) 가파르게 감소하고 수렴하는 값은 크다
- 밑수가 분수이면서 크면 (=지수가 정수이고 - 승이면) 완만하게 감소하고 수렴하는 값은 작다

```python
fig, ax = plt.subplots(1 ,2, figsize = (10, 5))

ax[0].plot(x1, y11, label=r'$\log_{10} x$', color='k')
ax[0].plot(x2, y12, '--', label=r'$\log_{e} x$', color='k')

ax[0].set_xlabel('$x$', fontsize=10)
ax[0].set_ylabel('$y$', fontsize=10)
ax[0].legend(fontsize=10, loc='lower right')

ax[1].plot(x1, y21, label=r'$\log_{1/10} x$', color='k')
ax[1].plot(x2, y22, '--', label=r'$\log_{1/e} x$', color='k')

ax[1].set_xlabel('$x$', fontsize=10)
ax[1].set_ylabel('$y$', fontsize=10)
ax[1].legend(fontsize=10, loc='upper right')

plt.show()
```

### 시그모이드(sigmoid) : $σ(z) = 1/1+e^{-z}$
- 값이 커지더라도 결국엔 1에 수렴하며 항상 0과 1사이의 값을 가진다

```python
z = np.linspace(-10, 10, 100)
sigma = 1/(1+np.exp(-z))

plt.figure(figsize=(10, 4))
plt.plot(z, sigma)
plt.xlabel('$z$', fontsize=10)
plt.ylabel('$\sigma(z)$', fontsize=10)
plt.show()
```

### 다변수 벡터함수를 위한 numpy의 meshgrid
- np.mashgrid(행방향 나열을 시키고 싶은 데이터, 열방향 나열을 시키고 싶은 데이터)
- 각각 데이터를 가진 2개의 변수를 행, 열에 나누고 이에 따른 순서로 모든 데이터를 연산한 데이터를 얻고자 할 때 사용

```python
t = np.linspace(0, 5, 6)
p = np.linspace(0, 5, 6)
```

```python
T, P = np.meshgrid(t, p)
# np.mashgrid(행방향 나열을 시키고 싶은 데이터, 열방향 나열을 시키고 싶은 데이터)
```

```python
T # T 를 행방향으로 데이터 나열
```

```python
P # P 를 열방향으로 데이터 나열
```

```python
T + P
# 두 변수의 모든 데이터를 행, 열 순서에 맞추어 다 더 한 경우의 수를 구할 수 있다
```

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)
U, V = np.meshgrid(u, v)
X = U
Y = V
Z = (1+U**2) + (V/(1+V**2))
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d') # 축을 3D로 그려라
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'$y$', fontsize=20)
ax.set_zlabel(r'$z$', fontsize=20)

ax.scatter3D(X, Y, Z, marker='.', color='gray')

plt.show()
```

### 함수의 합성

$x$ -> $y$ = $f(x)$ -> $z$ = $g(y)$

```python
x = np.linspace(-4, 4, 100)
y = x**3 - 15*x + 30 # f(x)
z = np.log(y) # g(y)
```

### 각각의 함수 표현

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y, label=r'$x^3 - 15x + 30$', color='k')
ax[0].legend(fontsize=18)

ax[1].plot(y, z, label=r'$\log(y)$', color='k')
ax[1].legend(fontsize=18)

plt.show()
```

### 함수 합성

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, z, '--', label=r'$\log(f(x))$', color='k')
ax[0].legend(fontsize=12)

ax[1].plot(x, y, label=r'$x^3 - 15x + 30$$', color='k')
ax[1].legend(fontsize=12, loc = 'upper right')
ax_tmp = ax[1].twinx()
ax_tmp.plot(x, z, '--', label = r'$\log(f(x))$', color='k')

plt.show()
```
