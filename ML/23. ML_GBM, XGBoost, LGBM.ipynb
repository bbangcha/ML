```python
import pandas as pd
url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/HAR_dataset/features.txt'

featuer_name_df = pd.read_csv(url, sep='\s+', 
                               # \s+ : 한개 이상의 띄어쓰기로 구분되어져 있다
                               header = None,
                               # header = : 인덱스를 이용하여 header 설정, none 시 없음
                               names = ['column_index', 'column_name'])
                               # name =[] : 불러올 때 칼럼 이름을 지정한 이름으로 변경
featuer_name = featuer_name_df.iloc[:, 1].values.tolist()
                                           # values : 값들만 추출
                                           # tolist() : 리스트로 변환
X_train_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/HAR_dataset/train/X_train.txt'
X_test_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/HAR_dataset/test/X_test.txt'

X_train = pd.read_csv(X_train_url, sep='\s+', header = None)
X_test = pd.read_csv(X_test_url, sep='\s+', header = None)
X_train.columns = featuer_name
X_test.columns = featuer_name
X_train.head()
```

```python
y_train_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/HAR_dataset/train/y_train.txt'
y_test_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/HAR_dataset/test/y_test.txt'

y_train = pd.read_csv(y_train_url, sep='\s+', header = None, names=['action'])
y_test = pd.read_csv(y_test_url, sep='\s+', header = None, names=['action'])

y_train.head()
```

### GBM
- 부스팅 알고리즘은 여러개의 약한 학습기(=매우 간단한 분류기)를 순차적으로 학습-예측
- 잘못 예측한 데이터에 가중치를 부여해서 오류를 개선하는 방식
- GBM은 가중치를 업데이트할 떄 경사 하강법(Gradient Descent)을 이용

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import time
import warnings

warnings.filterwarnings('ignore')
```

```python
start_time = time.time()
gb_clf = GradientBoostingClassifier(random_state=13)
gb_clf.fit(X_train, y_train)
gb_pred = gb_clf.predict(X_test)

print('ACC :', accuracy_score(y_test, gb_pred))
print('Fit time :', time.time() - start_time)
```

### Gridsearch 적용
- n_estimators : 사용할 DecisionTree 수
- learning_rate : 학습률로서 모델이 각 반복(또는 에폭)에서 가중치를 조절하는 정도
- 0과 1 사이의 실수이며 학습률이 작으면 정확하지만 시간이 오래걸리고, 학습률이 시간은 빠르지만 오류일 가능성이 높다
- verbose : 수행 결과 메시지를 출력
- 0 (기본값): 출력이 없음. 최적의 매개변수 및 교차 검증 결과만 반환됩니다.
- 1: 그리드 서치의 진행 상황에 대한 간단한 메시지를 출력합니다. 어떤 하이퍼파라미터 조합을 현재 평가하고 있는지 등의 기본 정보가 표시됩니다.
- 2: 상세한 출력을 제공합니다. 각각의 폴드에서 수행되는 교차 검증 결과 및 모든 조합에 대한 세부 정보를 출력합니다.

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators' : [100, 500],
    'learning_rate' : [0.05, 0.1]
}

start_time = time.time()
grid = GridSearchCV(estimator=gb_clf, param_grid=params, cv = 2, verbose=1, n_jobs = -1)
grid.fit(X_train, y_train)
print('Fit time :', time.time() - start_time)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

### Gridsearch의 best estimator test데이터 적용

```python
accuracy_score(y_test, grid.best_estimator_.predict(X_test))
```

### XGBoost
- 트리 기반의 앙상블 학습에서 각광받는 알고리즘 중 하나
- GBM 기반의 알고리즘을 다양한 규제를 통해 느린 속도 해결
- 병렬학습이 가능하도록 설계
- 반복 수행 시마다 내부적으로 학습데이터와 검증데이터를 교차검증 수행
- 교차검증을 통해 최적화가 완료되면 반복을 중단하는 조기 중단 기능 보유

### XGBoost 주요 파라미터
- nthread : CPU의 실행 스레드 개수를 조정, 디폴트는 CPU 전체 스레드 사용
- eta : GBM 학습률
- num_boost_rounds : n_estimators와 같은 기능
- max_depth

```python
from xgboost import XGBClassifier
```

### 조기 종료 조건과 검증 데이터 지정
- early_stopping_rounds = ? : 같은 성능으로 ?번 이상 반복해서 비슷한 성능이 나오면 종료
- eval_set : 검증set

```python
from xgboost import XGBClassifier

evals = [(X_test.values, y_test-1)]
# 원래는 종료조건을 걸기 위해서는 validation data를 넣어줘야되지만 따로 만들지 않고 test data로 대체하여 사용

start_time = time.time()
xgb = XGBClassifier(n_estimators = 400, learning_rate = 0.1, max_depth = 3)
xgb.fit(X_train.values, y_train-1, early_stopping_rounds = 10, eval_set = evals)
# sklean 모듈은 판다스 자료들을 잘 받아와 연동도 잘되지만 xgboost 모듈은 numpy array형태로 넣어줘야 하므로 .values로 변환해줘야 된다

print('Fit time :', time.time() - start_time)
```

```python
accuracy_score(y_test-1, xgb.predict(X_test.values))
```

### LightGBM

- LightGBM은 XGBoot와 함께 부스팅 계열에서 가장 각광받는 알고리즘
- LBGM의 가장 큰 장점은 속도
- 단, 적은 수의 데이터에는 어울이지 않음(일반적으로 10,000건 이상의 데이터가 필요)
- GPU 버전도 존재
- eval_set : 검증set
- callbacks=[early_stopping(?)] : 같은 성능으로 ?번 이상 반복해서 비슷한 성능이 나오면 종료(from lightgbm import early_stopping 필수)

```python
from lightgbm import LGBMClassifier
from lightgbm import early_stopping

evals = [(X_test.values, y_test)]
# 원래는 종료조건을 걸기 위해서는 validation data를 넣어줘야되지만 따로 만들지 않고 test data로 대체하여 사용

start_time = time.time()
lgbm = LGBMClassifier(n_estimators=400)
lgbm.fit(X_train.values, y_train, eval_set = evals, callbacks=[early_stopping(20)])
# sklean 모듈은 판다스 자료들을 잘 받아와 연동도 잘되지만 LGBM 모듈은 numpy array형태로 넣어줘야 하므로 .values로 변환해줘야 된다

print('Fit time :', time.time() - start_time)
```

```python
accuracy_score(y_test, lgbm.predict(X_test.values))
```
